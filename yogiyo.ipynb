```python
import pandas as pd
import numpy as np
```

```python
path = 'C:/work/pdata/'
ygy = pd.read_csv(path+'yogiyo_seoul.csv')
```

```python
chuga = pd.DataFrame({'광역시도':[],'시군구':[]})
```

```python
chuga
```

```python
yogiyo=pd.concat([chuga,ygy],axis=1)
yogiyo
```

```python
# 지역구로 구분
yogiyo['광역시도'] = yogiyo.address.str.split(' ').str[0]
yogiyo['시군구'] = yogiyo.address.str.split(' ').str[1]
yogiyo
```

```python
yogiyo.columns
```

```python
#피벗테이블
p_yogiyo=pd.pivot_table(yogiyo,index=['광역시도','시군구'],
               values=['delivery_fee','estimated_delivery_time_key','min_order_amount','review_count','review_avg','average_delivery','average_quantity','average_taste','review_reply_count'],
               aggfunc=[np.mean])
```

# 시각화

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

x= np.linspace(0,14,100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3)
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3)
plt.show()
```

```python
# matplotlib에서 한글 사용
# plt의 기본 글꼴은 영어 -> 한글 사용시 한글가능 폰트로 기본 폰트 변경 필요

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False #폰트 한글 변경 후 마이너스 기호 깨짐현상 극복

f_path = "C:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font',family=font_name)
```

```python
#박스플랏
plt.figure(figsize=(20,10))
sns.boxplot(x='광역시도',y='delivery_fee',data=yogiyo, hue='advertisement_rank',palette="Set3")
plt.show()
```

```python
#히트맵
plt.figure(figsize=(10,10))
sns.heatmap(p_yogiyo, annot=True,fmt='d')
plt.show()
```

```python
#pair plot
sns.pairplot(yogiyo,hue="additional_discount_pickup_only_for_user")
plt.show()
```
