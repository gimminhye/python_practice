# 이미지 마스킹

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

# 이미지 마스킹처리

```python
image_path = path+'korea_map.jpg'
korea_map = np.array(Image.open(image_path))
```

# 워드 클라우드 이미지 생성

```python
under_wordcloud = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',
                     background_color='white',
                     width=5000,
                     height=5000,
                     mask=korea_map).generate_from_frequencies(dict(under_tags))
```

# 화면출력

```python
fig = plt.figure()
plt.imshow(under_wordcloud, interpolation='bilinear', cmap='YlOrBr')
plt.axis('off')
# svg 객체로 이미지 저장하기
plt.savefig(path+'korea_upper_review.svg')
```
