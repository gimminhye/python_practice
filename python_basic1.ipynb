```python
df1 = pd.DataFrame({'만족도평점':[],'배달만족도':[],'음식양만족도':[],'맛만족도':[],'재방문의사':[]})
df1
```

```python
df1['만족도평점'] = ['4초과5미만','3초과4미만','2초과3미만','1초과2미만','1미만'] 
df1
```

```python
b = np.array(newrst['배달만족도'])
y = np.array(newrst['음식양만족도'])
m = np.array(newrst['맛만족도'])

for i in range(len(newrst)):
    if newrst['만족도평점'][i]>4:
        df1['배달만족도'][0] = b[np.where(newrst['만족도평점'][i])].mean()
        df1['음식양만족도'][0] = y[np.where(newrst['만족도평점'][i])].mean()
        df1['맛만족도'][0] = m[np.where(newrst['만족도평점'])].mean()
        df1['재방문의사'][0] = sum(newrst['재방문의사']=='그렇다')
        
    elif newrst['만족도평점'][i]>3 and newrst['만족도평점'][i]<4:
        df1['배달만족도'][1] = b[np.where(newrst['만족도평점'][i])].mean()
        df1['음식양만족도'][1] = y[np.where(newrst['만족도평점'][i])].mean()
        df1['맛만족도'][1] = m[np.where(newrst['만족도평점'])].mean()
        df1['재방문의사'][1] = sum(newrst['만족도평점']=='그렇다')

    elif newrst['만족도평점'][i]>2 and newrst['만족도평점'][i]<3:
        df1['배달만족도'][2] = b[np.where(newrst['만족도평점'][i])].mean()
        df1['음식양만족도'][2] = y[np.where(newrst['만족도평점'][i])].mean()
        df1['맛만족도'][2] = m[np.where(newrst['만족도평점'])].mean()
        df1['재방문의사'][2] = sum(newrst['만족도평점']=='그렇다')

    elif newrst['만족도평점'][i]>1 and newrst['만족도평점'][i]<2:
        df1['배달만족도'][3] = b[np.where(newrst['만족도평점'][i])].mean()
        df1['음식양만족도'][3] = y[np.where(newrst['만족도평점'][i])].mean()
        df1['맛만족도'][3] = m[np.where(newrst['만족도평점'])].mean()
        df1['재방문의사'][3] = sum(newrst['만족도평점']=='그렇다')

    else:
        df1['배달만족도'][4] = b[np.where(newrst['배달만족도'][i])].mean()
        df1['음식양만족도'][4] = y[np.where(newrst['음식양만족도'][i])].mean()
        df1['맛만족도'][4] = m[np.where(newrst['맛만족도'])].mean()
        df1['재방문의사'][4] = sum(newrst['재방문의사']=='그렇다')
df1
```
